```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating investment data for a portfolio over 12 months
months = pd.date_range(start='2023-01-01', periods=12, freq='M')
initial_investment = 100000  # Initial investment in dollars

# Simulating monthly returns for the portfolio and benchmark
portfolio_returns = np.random.normal(loc=0.01, scale=0.05, size=12)  # Monthly returns for the portfolio
benchmark_returns = np.random.normal(loc=0.008, scale=0.03, size=12)  # Monthly returns for the benchmark

# Calculating cumulative returns
portfolio_values = initial_investment * (1 + portfolio_returns).cumprod()
benchmark_values = initial_investment * (1 + benchmark_returns).cumprod()

# Creating a DataFrame to hold the results
performance_data = pd.DataFrame({
    'Month': months,
    'Portfolio_Value': portfolio_values,
    'Benchmark_Value': benchmark_values
})

# Calculating performance metrics
performance_data['Portfolio_Return'] = performance_data['Portfolio_Value'].pct_change() * 100
performance_data['Benchmark_Return'] = performance_data['Benchmark_Value'].pct_change() * 100
performance_data['Excess_Return'] = performance_data['Portfolio_Return'] - performance_data['Benchmark_Return']

# Displaying the performance data
print(performance_data)

# Visualizing Portfolio vs Benchmark Performance
plt.figure(figsize=(14, 6))
sns.lineplot(data=performance_data, x='Month', y='Portfolio_Value', label='Portfolio Value', marker='o', color='blue')
sns.lineplot(data=performance_data, x='Month', y='Benchmark_Value', label='Benchmark Value', marker='o', color='red')
plt.title('Portfolio Value vs Benchmark Value Over Time')
plt.xlabel('Month')
plt.ylabel('Value ($)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Visualizing Monthly Returns
plt.figure(figsize=(14, 6))
sns.lineplot(data=performance_data, x='Month', y='Portfolio_Return', label='Portfolio Return (%)', marker='o', color='blue')
sns.lineplot(data=performance_data, x='Month', y='Benchmark_Return', label='Benchmark Return (%)', marker='o', color='red')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Monthly Returns: Portfolio vs Benchmark')
plt.xlabel('Month')
plt.ylabel('Return (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Visualizing Excess Returns
plt.figure(figsize=(14, 6))
sns.barplot(data=performance_data, x='Month', y='Excess_Return', palette='Blues')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Excess Returns Over Time')
plt.xlabel('Month')
plt.ylabel('Excess Return (%)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
