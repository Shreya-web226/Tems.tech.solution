```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulate 12 months of data for finance sector
months = pd.date_range(start="2023-01-01", periods=12, freq='M')

# Simulate marketing and sales costs (in USD)
marketing_cost = np.random.randint(30000, 70000, size=12)
sales_cost = np.random.randint(15000, 40000, size=12)

# Simulate number of customers acquired through different channels
customers_acquired = np.random.randint(200, 800, size=12)

# Channel-specific customer data
channel_google_ads = customers_acquired * np.random.uniform(0.4, 0.6, size=12)
channel_referrals = customers_acquired * np.random.uniform(0.1, 0.3, size=12)
channel_social_media = customers_acquired * np.random.uniform(0.1, 0.3, size=12)

# Total cost and CAC calculation
total_cost = marketing_cost + sales_cost
CAC = total_cost / customers_acquired

# Create DataFrame
df_finance = pd.DataFrame({
    'Month': months,
    'Marketing Cost': marketing_cost,
    'Sales Cost': sales_cost,
    'Total Cost': total_cost,
    'Customers Acquired': customers_acquired,
    'CAC': CAC,
    'Google Ads Customers': channel_google_ads,
    'Referral Customers': channel_referrals,
    'Social Media Customers': channel_social_media
})

df_finance.head()  # Display first few rows
```

```python
# Plotting CAC trends over time by acquisition channels

plt.figure(figsize=(10, 6))

# Plot CAC over time
plt.plot(df_finance['Month'], df_finance['CAC'], marker='o', label='Overall CAC ($)', color='blue')

# Plot customers acquired through each channel
plt.plot(df_finance['Month'], df_finance['Google Ads Customers'], marker='o', label='Google Ads Customers', linestyle='--', color='green')
plt.plot(df_finance['Month'], df_finance['Referral Customers'], marker='o', label='Referral Customers', linestyle='--', color='orange')
plt.plot(df_finance['Month'], df_finance['Social Media Customers'], marker='o', label='Social Media Customers', linestyle='--', color='purple')

# Customize plot
plt.title('Customer Acquisition Cost (CAC) and Acquisitions by Channel')
plt.xlabel('Month')
plt.ylabel('Number of Customers / CAC ($)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
