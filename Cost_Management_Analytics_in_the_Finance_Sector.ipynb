```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating cost data for 12 months
months = pd.date_range(start='2023-01-01', periods=12, freq='M')
fixed_costs = np.random.normal(loc=30000, scale=2000, size=12).clip(0, None)  # Fixed costs
variable_costs = np.random.normal(loc=10000, scale=1000, size=12).clip(0, None)  # Variable costs
total_costs = fixed_costs + variable_costs  # Total costs

# Creating a DataFrame
cost_data = pd.DataFrame({
    'Month': months,
    'Fixed_Costs': fixed_costs,
    'Variable_Costs': variable_costs,
    'Total_Costs': total_costs
})

# Calculating cost proportions
cost_data['Fixed_Cost_Proportion'] = (cost_data['Fixed_Costs'] / cost_data['Total_Costs']) * 100
cost_data['Variable_Cost_Proportion'] = (cost_data['Variable_Costs'] / cost_data['Total_Costs']) * 100

# Display the cost data
print(cost_data)

# Visualizing Cost Breakdown
plt.figure(figsize=(14, 6))
sns.lineplot(data=cost_data, x='Month', y='Fixed_Costs', marker='o', label='Fixed Costs', color='blue')
sns.lineplot(data=cost_data, x='Month', y='Variable_Costs', marker='o', label='Variable Costs', color='red')
sns.lineplot(data=cost_data, x='Month', y='Total_Costs', marker='o', label='Total Costs', color='green')
plt.title('Cost Breakdown Over Time')
plt.xlabel('Month')
plt.ylabel('Cost Amount ($)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Visualizing Cost Proportions
plt.figure(figsize=(14, 6))
sns.barplot(data=cost_data, x='Month', y='Fixed_Cost_Proportion', color='blue', label='Fixed Cost Proportion', alpha=0.6)
sns.barplot(data=cost_data, x='Month', y='Variable_Cost_Proportion', color='red', label='Variable Cost Proportion', alpha=0.6)
plt.title('Cost Proportions Over Time')
plt.xlabel('Month')
plt.ylabel('Cost Proportion (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```
