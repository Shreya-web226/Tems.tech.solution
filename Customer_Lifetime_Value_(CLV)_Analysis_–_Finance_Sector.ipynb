```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating customer data
num_customers = 1000

# Simulating average purchase value (APV), purchase frequency (PF), and customer lifespan (CL)
average_purchase_value = np.random.normal(loc=150, scale=30, size=num_customers)  # APV in dollars
purchase_frequency = np.random.poisson(lam=5, size=num_customers)  # PF in purchases per year
customer_lifespan = np.random.randint(low=1, high=11, size=num_customers)  # CL in years

# Creating a DataFrame to hold customer data
customer_data = pd.DataFrame({
    'Customer_ID': np.arange(1, num_customers + 1),
    'Average_Purchase_Value': average_purchase_value,
    'Purchase_Frequency': purchase_frequency,
    'Customer_Lifespan': customer_lifespan
})

# Calculating Customer Lifetime Value (CLV)
customer_data['CLV'] = customer_data['Average_Purchase_Value'] * \
                        customer_data['Purchase_Frequency'] * \
                        customer_data['Customer_Lifespan']

# Displaying the first few rows of customer data
print(customer_data.head())

# Visualizing the distribution of Customer Lifetime Values
plt.figure(figsize=(14, 6))
sns.histplot(customer_data['CLV'], bins=30, kde=True, color='blue')
plt.title('Distribution of Customer Lifetime Value (CLV)')
plt.xlabel('Customer Lifetime Value ($)')
plt.ylabel('Frequency')
plt.grid()
plt.show()

# Visualizing CLV against Purchase Frequency
plt.figure(figsize=(14, 6))
sns.scatterplot(data=customer_data, x='Purchase_Frequency', y='CLV', color='green')
plt.title('CLV vs. Purchase Frequency')
plt.xlabel('Purchase Frequency (Purchases/Year)')
plt.ylabel('Customer Lifetime Value ($)')
plt.grid()
plt.show()

# Visualizing CLV against Customer Lifespan
plt.figure(figsize=(14, 6))
sns.boxplot(data=customer_data, x='Customer_Lifespan', y='CLV', palette='Pastel1')
plt.title('CLV Distribution by Customer Lifespan')
plt.xlabel('Customer Lifespan (Years)')
plt.ylabel('Customer Lifetime Value ($)')
plt.grid()
plt.show()
```
