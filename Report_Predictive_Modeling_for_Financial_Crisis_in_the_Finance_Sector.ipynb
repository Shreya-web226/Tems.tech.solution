```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure to hold multiple subplots
plt.figure(figsize=(15, 10))

# Subplot 1: GDP Growth Distribution
plt.subplot(2, 2, 1)
sns.histplot(df['GDP_Growth'], kde=True, color='blue')
plt.title('GDP Growth Distribution')
plt.xlabel('GDP Growth (%)')

# Subplot 2: Inflation Rate Distribution
plt.subplot(2, 2, 2)
sns.histplot(df['Inflation_Rate'], kde=True, color='green')
plt.title('Inflation Rate Distribution')
plt.xlabel('Inflation Rate (%)')

# Subplot 3: Interest Rate Distribution
plt.subplot(2, 2, 3)
sns.histplot(df['Interest_Rate'], kde=True, color='red')
plt.title('Interest Rate Distribution')
plt.xlabel('Interest Rate (%)')

# Subplot 4: Unemployment Rate vs. Financial Crisis
plt.subplot(2, 2, 4)
sns.boxplot(x='Financial_Crisis', y='Unemployment_Rate', data=df)
plt.title('Unemployment Rate vs Financial Crisis')
plt.xlabel('Financial Crisis (0 = No, 1 = Yes)')
plt.ylabel('Unemployment Rate (%)')

plt.tight_layout()
plt.show()
```
