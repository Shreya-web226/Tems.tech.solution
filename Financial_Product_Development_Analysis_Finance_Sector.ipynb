```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate synthetic financial product data
np.random.seed(42)

# Time period (months)
months = 36

# Product categories: Mutual Funds, Insurance, Loans
products = ['Mutual Funds', 'Insurance', 'Loans']

# Simulated product return rates over time (%)
returns = {
    'Mutual Funds': np.random.normal(0.05, 0.02, months),
    'Insurance': np.random.normal(0.03, 0.01, months),
    'Loans': np.random.normal(0.06, 0.03, months)
}

# Simulated risk level (standard deviation) for each product over time
risk = {
    'Mutual Funds': np.random.normal(0.02, 0.01, months),
    'Insurance': np.random.normal(0.01, 0.005, months),
    'Loans': np.random.normal(0.03, 0.015, months)
}

# Customer acquisition rate (customers per month)
customers = {
    'Mutual Funds': np.random.poisson(100, months),
    'Insurance': np.random.poisson(80, months),
    'Loans': np.random.poisson(120, months)
}

# Step 2: Create DataFrame for analysis
data = pd.DataFrame({
    'Month': np.arange(1, months + 1),
    'Mutual Fund Returns': returns['Mutual Funds'],
    'Insurance Returns': returns['Insurance'],
    'Loan Returns': returns['Loans'],
    'Mutual Fund Risk': risk['Mutual Funds'],
    'Insurance Risk': risk['Insurance'],
    'Loan Risk': risk['Loans'],
    'Mutual Fund Customers': customers['Mutual Funds'],
    'Insurance Customers': customers['Insurance'],
    'Loan Customers': customers['Loans']
})

# Step 3: Basic Analysis
# 3.1 Correlation Analysis - To find relationships between returns and risk for each product
correlation_matrix = data[['Mutual Fund Returns', 'Insurance Returns', 'Loan Returns',
                           'Mutual Fund Risk', 'Insurance Risk', 'Loan Risk']].corr()

print("Correlation Matrix:\n", correlation_matrix)

# Step 4: Risk-Return Profile Plot
plt.figure(figsize=(10, 6))

plt.scatter(data['Mutual Fund Risk'], data['Mutual Fund Returns'], color='blue', label='Mutual Funds')
plt.scatter(data['Insurance Risk'], data['Insurance Returns'], color='green', label='Insurance')
plt.scatter(data['Loan Risk'], data['Loan Returns'], color='red', label='Loans')

plt.title('Risk-Return Profile of Financial Products')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Returns (%)')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Customer Growth Trends
plt.figure(figsize=(10, 6))

plt.plot(data['Month'], data['Mutual Fund Customers'], label='Mutual Funds')
plt.plot(data['Month'], data['Insurance Customers'], label='Insurance')
plt.plot(data['Month'], data['Loan Customers'], label='Loans')

plt.title('Customer Acquisition Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Summary statistics for the performance of each product
summary_stats = data.describe()

print("Summary Statistics:\n", summary_stats)
```
