```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating product data
num_products = 10

# Generating product names
product_names = [f'Product_{i+1}' for i in range(num_products)]

# Simulating revenue and costs
revenue = np.random.normal(loc=200000, scale=50000, size=num_products).clip(0, None)  # Revenue in dollars
costs = np.random.normal(loc=150000, scale=30000, size=num_products).clip(0, None)  # Costs in dollars

# Creating a DataFrame to hold product data
product_data = pd.DataFrame({
    'Product_Name': product_names,
    'Revenue': revenue,
    'Costs': costs
})

# Calculating Profit and Profit Margin
product_data['Profit'] = product_data['Revenue'] - product_data['Costs']
product_data['Profit_Margin'] = (product_data['Profit'] / product_data['Revenue']) * 100

# Displaying the product data
print(product_data)

# Visualizing Revenue, Costs, and Profit for Each Product
plt.figure(figsize=(14, 6))
bar_width = 0.25
index = np.arange(num_products)

# Creating bar plots for Revenue, Costs, and Profit
plt.bar(index, product_data['Revenue'], bar_width, label='Revenue', color='blue')
plt.bar(index + bar_width, product_data['Costs'], bar_width, label='Costs', color='orange')
plt.bar(index + 2 * bar_width, product_data['Profit'], bar_width, label='Profit', color='green')

plt.xlabel('Products')
plt.ylabel('Amount ($)')
plt.title('Revenue, Costs, and Profit for Each Product')
plt.xticks(index + bar_width, product_data['Product_Name'], rotation=45)
plt.legend()
plt.grid()
plt.show()

# Visualizing Profit Margins
plt.figure(figsize=(14, 6))
sns.barplot(data=product_data, x='Product_Name', y='Profit_Margin', palette='Blues')
plt.axhline(y=0, color='black', linewidth=0.8, linestyle='--')
plt.title('Profit Margin for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Analyzing and Visualizing Profitability
# Highlighting profitable vs unprofitable products
product_data['Profitability_Status'] = np.where(product_data['Profit'] > 0, 'Profitable', 'Unprofitable')

plt.figure(figsize=(14, 6))
sns.countplot(data=product_data, x='Profitability_Status', palette='pastel')
plt.title('Number of Profitable vs Unprofitable Products')
plt.xlabel('Profitability Status')
plt.ylabel('Count')
plt.grid()
plt.show()
```
