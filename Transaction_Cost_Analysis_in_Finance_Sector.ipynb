```python
import numpy as np
import random

# Simulate trade data for a hypothetical security
# Number of trades
n_trades = 100

# Generate random data for each trade
# Simulated stock prices (in dollars)
prices = np.round(np.random.uniform(50, 150, n_trades), 2)

# Simulated volumes (in shares)
volumes = np.random.randint(100, 1000, n_trades)

# Desired trade execution price (mid-market reference price)
reference_price = 100

# Calculate VWAP (Volume-Weighted Average Price)
def calculate_vwap(prices, volumes):
    vwap = np.sum(prices * volumes) / np.sum(volumes)
    return vwap

# Calculate Implementation Shortfall
def calculate_implementation_shortfall(prices, volumes, reference_price):
    # Calculate actual transaction cost as difference from reference price
    executed_cost = np.sum(prices * volumes)
    ideal_cost = reference_price * np.sum(volumes)

    # Shortfall as a percentage of the ideal cost
    implementation_shortfall = (executed_cost - ideal_cost) / ideal_cost * 100
    return implementation_shortfall

# Compute results
vwap = calculate_vwap(prices, volumes)
implementation_shortfall = calculate_implementation_shortfall(prices, volumes, reference_price)

# Output results
print("Transaction Cost Analysis Metrics:")
print(f"Volume-Weighted Average Price (VWAP): ${vwap:.2f}")
print(f"Implementation Shortfall: {implementation_shortfall:.2f}%")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up simulated trade data
n_trades = 100
prices = np.round(np.random.uniform(50, 150, n_trades), 2)
volumes = np.random.randint(100, 1000, n_trades)
reference_price = 100

# VWAP Calculation
def calculate_vwap(prices, volumes):
    vwap = np.sum(prices * volumes) / np.sum(volumes)
    return vwap

# Implementation Shortfall Calculation
def calculate_implementation_shortfall(prices, volumes, reference_price):
    executed_cost = np.sum(prices * volumes)
    ideal_cost = reference_price * np.sum(volumes)
    implementation_shortfall = (executed_cost - ideal_cost) / ideal_cost * 100
    return implementation_shortfall

# Compute TCA metrics
vwap = calculate_vwap(prices, volumes)
implementation_shortfall = calculate_implementation_shortfall(prices, volumes, reference_price)

# Plot Trade Prices and Volumes
plt.figure(figsize=(12, 6))

# Plot Prices
plt.subplot(1, 2, 1)
plt.plot(prices, color='blue', marker='o', label='Trade Prices')
plt.axhline(y=vwap, color='green', linestyle='--', label=f'VWAP: ${vwap:.2f}')
plt.axhline(y=reference_price, color='red', linestyle='--', label=f'Reference Price: ${reference_price}')
plt.title("Trade Prices with VWAP and Reference Price")
plt.xlabel("Trade Number")
plt.ylabel("Price ($)")
plt.legend()

# Plot Volumes
plt.subplot(1, 2, 2)
plt.bar(range(n_trades), volumes, color='purple', alpha=0.6)
plt.title("Trade Volumes")
plt.xlabel("Trade Number")
plt.ylabel("Volume (Shares)")

plt.tight_layout()
plt.show()

# Summary of Metrics
print("Transaction Cost Analysis Metrics:")
print(f"Volume-Weighted Average Price (VWAP): ${vwap:.2f}")
print(f"Implementation Shortfall: {implementation_shortfall:.2f}%")
```
