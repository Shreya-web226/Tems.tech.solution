```python
import numpy as np
import matplotlib.pyplot as plt

n_loans = 1000
loan_amount = 100000
pd = 0.02
lgd = 0.5


n_simulations = 100000


def simulate_losses(n_loans, loan_amount, pd, lgd, n_simulations):
    losses = np.zeros(n_simulations)
    for i in range(n_simulations):

        defaults = np.random.binomial(1, pd, n_loans)
        losses[i] = np.sum(defaults * loan_amount * lgd)
    return losses


losses = simulate_losses(n_loans, loan_amount, pd, lgd, n_simulations)

economic_capital = np.percentile(losses, 99.5)
print(f"Economic Capital at 99.5% confidence level: ${economic_capital:,.2f}")
plt.figure(figsize=(10, 6))
plt.hist(losses, bins=50, color='blue', alpha=0.7)
plt.axvline(economic_capital, color='red', linestyle='dashed', linewidth=2)
plt.title('Distribution of Simulated Losses')
plt.xlabel('Loss Amount ($)')
plt.ylabel('Frequency')
plt.show()
```
