```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating revenue and cost data for 12 months
months = pd.date_range(start='2023-01-01', periods=12, freq='M')
interest_income = np.random.normal(loc=50000, scale=5000, size=12).clip(0, None)
fee_income = np.random.normal(loc=10000, scale=2000, size=12).clip(0, None)
commission_income = np.random.normal(loc=15000, scale=3000, size=12).clip(0, None)
total_revenue = interest_income + fee_income + commission_income

fixed_costs = np.random.normal(loc=30000, scale=2000, size=12).clip(0, None)
variable_costs = np.random.normal(loc=10000, scale=1000, size=12).clip(0, None)
total_costs = fixed_costs + variable_costs

# Creating a DataFrame
financial_data = pd.DataFrame({
    'Month': months,
    'Interest_Income': interest_income,
    'Fee_Income': fee_income,
    'Commission_Income': commission_income,
    'Total_Revenue': total_revenue,
    'Total_Costs': total_costs
})

# Calculating Profitability Metrics
financial_data['Gross_Profit'] = financial_data['Total_Revenue'] - financial_data['Total_Costs']
financial_data['Gross_Profit_Margin'] = (financial_data['Gross_Profit'] / financial_data['Total_Revenue']) * 100
financial_data['Net_Profit_Margin'] = (financial_data['Gross_Profit'] / financial_data['Total_Revenue']) * 100

# Display the financial data
print(financial_data)

# Visualizing Revenue and Costs
plt.figure(figsize=(14, 6))
sns.lineplot(data=financial_data, x='Month', y='Total_Revenue', label='Total Revenue', marker='o', color='blue')
sns.lineplot(data=financial_data, x='Month', y='Total_Costs', label='Total Costs', marker='o', color='red')
plt.title('Total Revenue and Costs Over Time')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Visualizing Profit Margins
plt.figure(figsize=(14, 6))
sns.lineplot(data=financial_data, x='Month', y='Gross_Profit_Margin', marker='o', color='green', label='Gross Profit Margin')
sns.lineplot(data=financial_data, x='Month', y='Net_Profit_Margin', marker='o', color='purple', label='Net Profit Margin')
plt.title('Profit Margins Over Time')
plt.xlabel('Month')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```
