```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a random seed for reproducibility
np.random.seed(42)

# Simulating a portfolio
initial_portfolio_value = 1_000_000  # Initial portfolio value in dollars
portfolio = {
    'Stocks': 0.6 * initial_portfolio_value,    # 60% in stocks
    'Bonds': 0.3 * initial_portfolio_value,     # 30% in bonds
    'Cash': 0.1 * initial_portfolio_value        # 10% in cash
}

# Defining stress scenarios (percentage change in portfolio value)
scenarios = {
    'Economic Downturn': -0.30,  # 30% drop
    'Interest Rate Hike': -0.15,  # 15% drop
    'Market Crash': -0.50,        # 50% drop
    'Inflation Spike': -0.10       # 10% drop
}

# Creating a list to hold scenario results
scenario_data = []

# Calculating impacts of each scenario and appending to the list
for scenario, impact in scenarios.items():
    new_value = initial_portfolio_value * (1 + impact)
    scenario_data.append({
        'Scenario': scenario,
        'Impact (%)': impact * 100,
        'New Portfolio Value': new_value
    })

# Creating a DataFrame from the list of scenario data
scenario_analysis = pd.DataFrame(scenario_data)

# Displaying the scenario analysis results
print(scenario_analysis)

# Visualizing Scenario Analysis
plt.figure(figsize=(10, 6))
sns.barplot(data=scenario_analysis, x='Scenario', y='Impact (%)', palette='Blues')
plt.title('Impact of Stress Scenarios on Portfolio Value')
plt.xlabel('Scenario')
plt.ylabel('Impact (%)')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(axis='y')
plt.show()

# Visualizing New Portfolio Values
plt.figure(figsize=(10, 6))
sns.barplot(data=scenario_analysis, x='Scenario', y='New Portfolio Value', palette='Reds')
plt.title('New Portfolio Value Under Stress Scenarios')
plt.xlabel('Scenario')
plt.ylabel('New Portfolio Value ($)')
plt.axhline(initial_portfolio_value, color='black', linewidth=0.8, linestyle='--', label='Initial Portfolio Value')
plt.legend()
plt.grid(axis='y')
plt.show()
```
